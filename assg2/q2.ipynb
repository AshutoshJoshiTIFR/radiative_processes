```python
import numpy as np
import matplotlib.pyplot as plt

# Given parameters
n_i = 50 * 1e6                    # m^-3
L = 1.0                           # m
sigma_T = 6.65e-29                # m^2
delta = 1/100                     # omega0 * reaction time scale

# Frequency ratio x = omega/omega0
x = np.logspace(-4, 2, 1000)

# Optical depth tau(x)
tau = n_i * L * sigma_T * x**4 / ((x**2 - 1)**2 + delta**2)

# Plotting
plt.figure(figsize=(8, 6))
plt.loglog(x, tau)
plt.xlabel(r'$\omega / \omega_0$')
plt.ylabel(r'Optical depth $\tau(\omega)$')
plt.grid(True, which='both', ls='--')
plt.show()
```
